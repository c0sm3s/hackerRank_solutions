```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
class Editor:
    def __init__(self, s=''):
        self.s = s
        self.temp = s
        
    def my_append(self, w): # 1
        self.temp = self.s
        self.s += w
    
    def my_delete(self, k): # 2
        self.temp = self.s
        self.s = self.s[:-k]
    
    def my_print(self, k): # 3
        print(self.s[k-1])
        
    def my_undo(self):
        self.s = self.temp
    
        
# if __name__ == "__main__":
my_editor = Editor()
q = int(input())
while q > 0:
    i = input().split()
    if i[0] == '1': my_editor.my_append(i[1])
    if i[0] == '2': my_editor.my_delete(int(i[1]))
    if i[0] == '3': my_editor.my_print(int(i[1]))
    if i[0] == '4': my_editor.my_undo()
    q -= 1
            
            
        
        
        
      
```

```python
def extraLongFactorials(n):
    # Write your code here
    i = 1
    tmp = 1
    while(i < n):
        tmp = tmp * (i+1)
        # print(tmp)
        i += 1
    return tmp

print(extraLongFactorials(5))
```
