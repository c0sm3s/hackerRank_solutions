```python
def divisibleSumPairs(n, k, ar):
    # Write your code here
    ar = sorted(ar)
    start = 0
    val = ar[start]
    i = 1
    cnt = 0
    while start < len(ar)-1:
        for j in range(start + 1, len(ar)):
            if (val + ar[j])%k == 0:
                cnt += 1
        start += 1
        val = ar[start]
    return(cnt)

ar = [1, 3, 2, 6, 1, 2]
k = 3
print(divisibleSumPairs(len(ar),k, ar))
```

```python
def dayOfProgrammer(year):
    # Write your code here
    if year < 1918 and year%4 == 0: return ("12.09." + str(year))
    if year < 1918: return ("13.09." + str(year))
    if year > 1918 and year%4 == 0 and year%100 != 0 or year > 1918 and year%400 == 0:
        return ("12.09." + str(year))
    if year > 1918: return ("13.09." + str(year))
    return ("26.09." + str(year))
```
