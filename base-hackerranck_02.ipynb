```python
def divisibleSumPairs(n, k, ar):
    # Write your code here
    ar = sorted(ar)
    start = 0
    val = ar[start]
    i = 1
    cnt = 0
    while start < len(ar)-1:
        for j in range(start + 1, len(ar)):
            if (val + ar[j])%k == 0:
                cnt += 1
        start += 1
        val = ar[start]
    return(cnt)

ar = [1, 3, 2, 6, 1, 2]
k = 3
print(divisibleSumPairs(len(ar),k, ar))
```
